```python
#!/usr/bin/env python

import cv2
import time

if __name__ == '__main__' :

    # Start default camera
    video = cv2.VideoCapture(0);

    # Find OpenCV version
    (major_ver, minor_ver, subminor_ver) = (cv2.__version__).split('.')

    # With webcam get(CV_CAP_PROP_FPS) does not work.
    # Let's see for ourselves.

    if int(major_ver)  < 3 :
        fps = video.get(cv2.cv.CV_CAP_PROP_FPS)
        print("Frames per second using video.get(cv2.cv.CV_CAP_PROP_FPS): {0}".format(fps))
    else :
        fps = video.get(cv2.CAP_PROP_FPS)
        print("Frames per second using video.get(cv2.CAP_PROP_FPS) : {0}".format(fps))

    # Number of frames to capture
    num_frames = 120;

    print("Capturing {0} frames".format(num_frames))

    # Start time
    start = time.time()

    # Grab a few frames
    for i in range(0, num_frames) :
        ret, frame = video.read()

    # End time
    end = time.time()

    # Time elapsed
    seconds = end - start
    print ("Time taken : {0} seconds".format(seconds))

    # Calculate frames per second
    fps  = num_frames / seconds
    print("Estimated frames per second : {0}".format(fps))

    # Release video
    video.release()
```

```python
import PySimpleGUI as sg            # Uncomment 1 to run on that framework
# import PySimpleGUIQt as sg        # runs on Qt with no other changes
# import PySimpleGUIWeb as sg       # has a known flicker problem that's being worked
import cv2
import mediapipe as mp
import imutils
import copy
import itertools
import csv
import numpy as np
from matplotlib import pyplot as plt
import math
import PIL
from PIL import Image,ImageTk
import cv2
from tkinter import *
import os
from vlc import Instance
import time
from math import atan2,degrees



class VLC:
    
    def __init__(self):
        self.Player = Instance('--loop')
#         self.Player.set_playback_mode(vlc.PlaybackMode.loop)

    def addPlaylist(self):
        self.mediaList = self.Player.media_list_new()
        s1='song1.mp3'
        s2='song2.mp3'
        s3='song3.mp3'
        s4='song4.mp3'
        os.chdir("C:\\Users\\farah\\Desktop\\music\\playlist\\")
        # Fetching Songs
#         songs = [s1,s2,s3]
        songs=os.listdir()
        
        for s in songs:
            self.mediaList.add_media(self.Player.media_new(s))
        self.listPlayer = self.Player.media_list_player_new()
        self.listPlayer.set_media_list(self.mediaList)
    def play(self):
        if self.listPlayer.is_playing()==0:
            self.listPlayer.play()
       
        window['PLAY'].update(image_filename='C:\\Users\\farah\\Desktop\\music\\pause.png')
    def next(self):
        self.listPlayer.next()
        
        window['NEXT'].update(image_filename='C:\\Users\\farah\\Desktop\\music\\nexton.png')
        window['PLAY'].update(image_filename='C:\\Users\\farah\\Desktop\\music\\pause.png')
        
        
    def pause(self):
        self.listPlayer.set_pause(1)
        window['PLAY'].update(image_filename='C:\\Users\\farah\\Desktop\\music\\play.png')
    def previous(self):
        self.listPlayer.previous()
       
        window['PREV'].update(image_filename='C:\\Users\\farah\\Desktop\\music\\prevon.png')
        window['PLAY'].update(image_filename='C:\\Users\\farah\\Desktop\\music\\pause.png')
    def get_time(self):
        return self.listplayer.get_time()
    def stop(self):
        self.listPlayer.stop()
    def setvol(self,v):
        self.listPlayer.get_media_player().audio_set_volume(v) 
        event, values = window.read(timeout=10)
        window['volume'].update(v)
    
    def clear(self):
        
        window['NEXT'].update(image_filename='C:\\Users\\farah\\Desktop\\music\\nextoff.png')
        window['PREV'].update(image_filename='C:\\Users\\farah\\Desktop\\music\\prevoff.png')
        
    def get_track_info(self):
        """ Show author, title, and elasped time if video is loaded and playing """
        time_elapsed = "{:02d}:{:02d}".format(*divmod(self.listPlayer.get_media_player().get_time() // 1000, 60))
        time_total = "{:02d}:{:02d}".format(*divmod(self.listPlayer.get_media_player().get_length() // 1000, 60))
        if self.listPlayer.is_playing():
            
            window['TIME_ELAPSED'].update(time_elapsed)
            if(self.listPlayer.get_media_player().get_length()!=0):
                window['TIME'].update(self.listPlayer.get_media_player().get_time()/self.listPlayer.get_media_player().get_length())
            window['TIME_TOTAL'].update(time_total)
            txt=self.listPlayer.get_media_player().get_media().get_mrl()
            x=txt.split('/')  
            
            window['songname'].update(x[9])


            
            
def calc_bounding_rect(image, landmarks):
    image_width, image_height = image.shape[1], image.shape[0]

    landmark_array = np.empty((0, 2), int)

    for _, landmark in enumerate(landmarks.landmark):
        landmark_x = min(int(landmark.x * image_width), image_width - 1)
        landmark_y = min(int(landmark.y * image_height), image_height - 1)

        landmark_point = [np.array((landmark_x, landmark_y))]

        landmark_array = np.append(landmark_array, landmark_point, axis=0)

    x, y, w, h = cv2.boundingRect(landmark_array)
    

    return [x, y, x + w, y + h,h]           
            
            
            
            
            
    
def is_clockwise(p):
    # points is your list (or array) of 2d points.
#     p=p[0]
    print(p)
#     print(p[0][1])
#     print(p[0][0])
#     print(p[1][1])
#     print(p[1][0])
    x=len(p)
    if(x==0):
        return -1
    s = 0.0
    x0=0
    y0=0
    t=[]
    oo=0
    for index in range(x):
        
        if(index>0):
                
#               oo= (p[index][0] - p[index-1][0]) * (p[index][1] + p[index-1][1])
#               s+= (p[index][0] - p[index-1][0]) * (p[index][1] + p[index-1][1])
#                 print(p[index][0][1])
#                 print(p[index-1][0][0])
#                 print(p[index-1][0][1])
#                 print(p[index][0][0])
                s+= (p[index][1] * p[index-1][0]) - (p[index][0] * p[index-1][1])
#               print(oo)
# #               s += (p[index][1] * p[index-1][0]) - (p[index][0] + p[index-1][1])
#               s=(p[index][0]*y0)-(x0*p[index][1])
#               s= (    (p[index-1][1] * p[index-1][0]) + (p[index-2][0] + p[index-1][1]) + (p[index-2][1]*p[index][0])  ) - (    (p[index-2][1]*p[index-1][0]) + (p[index-1][1]*p[index][0]) + (p[index-2][0]*p[index][1] )     )
       
           
        
#     print(t)
#     direction(t) 
    print(s)
    return s>=0


































def direction(arr):
    count=1
    o=arr
    t=[]
    p=[]
    x=len(o)
    for index in range(x):
      
        if(index>0):
           if(o[index-1]==o[index]):
             count=count+1
             
            
           else:
            t.append(count)
            p.append(o[index-1])
            count=1
        if(index==(x-1)):
             p.append(o[index])
             t.append(count)
            
    print(t)
    print(p)
    
#     max_value = max(t)



#     max_index = t.index(max_value) 
#     return p[max_index]
    
#     if(t[0]>t[1]):
#                      return p[0]
#     else:
#                      return p[1]








def getAngle(a_x, a_y, b_x, b_y):
        
             ab_x = b_x - a_x;
             ab_y = b_y - a_y;
#              angleDeg = math.atan2(ab_y,   ab_x) * 180 / math.pi;
             return degrees(atan2(ab_x,ab_y))
             #print(angleDeg);
#              return angleDeg   

        


def getAngle2(p1_x, p1_y, p2_x, p2_y):
        
        return math.acos((((p1_x * p2_x) + (p1_y * p2_y)) / (math.sqrt(math.pow(p1_x, 2) + math.pow(p1_y, 2)) * math.sqrt(math.pow(p2_x, 2) + math.pow(p2_y, 2))))) * 180 / math.pi;

        
def calc_landmark_list(image, landmarks):
    image_width, image_height = image.shape[1], image.shape[0]

    landmark_point = []

    # Keypoint
    for _, landmark in enumerate(landmarks.landmark):
        landmark_x = min(int(landmark.x * image_width), image_width - 1)
        landmark_y = min(int(landmark.y * image_height), image_height - 1)
        landmark_z = landmark.z

        landmark_point.append([landmark_x, landmark_y, landmark_z])

    return landmark_point


def DistanceAB(a_x,  a_y,  b_x,  b_y):
        
            dist = pow(a_x - b_x, 2) + pow(a_y - b_y, 2);
           
            return math.sqrt(dist);
        



def pre_process_landmark(landmark_list):
    temp_landmark_list = copy.deepcopy(landmark_list)

    # Convert to relative coordinates
    base_x, base_y = 0, 0
    for index, landmark_point in enumerate(temp_landmark_list):
        if index == 0:
            base_x, base_y = landmark_point[0], landmark_point[1]

        temp_landmark_list[index][0] = temp_landmark_list[index][0] - base_x
        temp_landmark_list[index][1] = temp_landmark_list[index][1] - base_y

    # Convert to a one-dimensional list
    temp_landmark_list = list(itertools.chain.from_iterable(temp_landmark_list))

    # Normalization
    max_value = max(list(map(abs, temp_landmark_list)))

    def normalize_(n):
        return n / max_value

    temp_landmark_list = list(map(normalize_, temp_landmark_list))

    return temp_landmark_list





# define the window layout

# Images are located in a subfolder in the Demo Media Player.py folder
image_pause = 'C:\\Users\\farah\\Desktop\\music\\play.png'
image_prev = 'C:\\Users\\farah\\Desktop\\music\\prevoff.png'
image_next = 'C:\\Users\\farah\\Desktop\\music\\nextoff.png'


# Use the theme APIs to set the buttons to blend with background
sg.theme_button_color((sg.theme_background_color(), sg.theme_background_color()))
sg.theme_border_width(0)        # make all element flat

col1=[ [sg.Button(image_filename=image_prev,key='PREV'),

          sg.Button(image_filename=image_pause,key='PLAY'),

          sg.Button(image_filename=image_next,key='NEXT'),

        ]]

col2=[ [sg.ProgressBar(100,orientation="vertical",size=(10,8),bar_color=['green','red'],key='volume', style='calm')]]

# define layout of the rows
layout= [
    
    [sg.Text('   ', key='songname', size=(50,2) , font=("Helvetica", 20), justification='center')],
    
       [sg.Text('00:00', key='TIME_ELAPSED'),

             # This slide can be used to adjust the playback positon of the video
             sg.Slider(range=(0, 1), default_value=0, size=(50, 20),enable_events=True, resolution=0.0001, disable_number_display=True,
                       background_color='#83D8F5', orientation='h', key='TIME'),

             # Elements for tracking media length and track counts
             sg.Text('00:00', key='TIME_TOTAL')],

             [sg.Column(col1), sg.Column(col2)]
            # Button and media information layout (created above)
           
    
        
   
         ]


# create the window and show it without the plot
window = sg.Window('Media Player', layout, element_justification='center')

# ---===--- Event LOOP Read and display frames, operate the GUI --- #
cap = cv2.VideoCapture(0)  
test=[]
test2=[]
points1=[]
points2=[]
order=[]
order2=[]
prev0=0
prev01=0
prevx=0
prevy=0
prevx2=0
prevx21=0
prevx22=0
prevx23=0
prevy2=0
prevx3=0
prevy3=0
dis=0
incno=0
stableno=0
stableno2=0
stableno3=0
stableno4=0
decno=0
incno2=0
decno2=0
first="true"
status="null"
too="NO"
ca=""
tto=""
di="NO"
font=5
com1=0
com2=0
first2="true"
prevangle=0
ori=[]
sa1=0
sa11=0
sa2=0

prevf1y=0
prevf2y=0



player = VLC()



player.addPlaylist()
v=50
player.setvol(v)



imgcanvas=np.zeros((480,640,3),np.uint8)
cap = cv2.VideoCapture(0)
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
cap = cv2.VideoCapture(0)
with mp_hands.Hands(
    min_detection_confidence=0.5,
    max_num_hands=1,
    min_tracking_confidence=0.5,
) as hands:
    while True:
        event, values = window.Read(timeout=20, timeout_key='timeout')
        if event is None:
            break
            
        if event == 'NEXT':
            player.next()    
        thumbIsOpen = "false"
        firstFingerIsOpen = "false"
        secondFingerIsOpen = "false"
        thirdFingerIsOpen = "false"
        fourthFingerIsOpen = "false"
        checkk="false"
        finger1="false"
        finger2="false"
        finger3="false"
        finger4="false"
        check2="false"
        check3="stable"
        
        
        success, image = cap.read()
#         print(image.shape)
        H ,N,NN=image.shape
#         print(H)
        player.clear()
        player.get_track_info()
#         image = imutils.resize(frame, width=320)

        image = cv2.flip(image, 1)  # Mirror display
        debug_image = copy.deepcopy(image)

            # Detection implementation #############################################################
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        results = hands.process(image)

        # Draw the hand annotations on the image.
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        #print("enterrrrrr")



        if results.multi_hand_landmarks:
          for hand_landmarks in results.multi_hand_landmarks:
             #print('da5allll ahoooo')
             
             #print(hand_landmarks.landmark)
#              test=hand_landmarks.landmark
             test2=hand_landmarks
             landmark = calc_landmark_list(debug_image, test2)
             brect = calc_bounding_rect(debug_image, hand_landmarks)
             
             mp_drawing.draw_landmarks(image, hand_landmarks, mp_hands.HAND_CONNECTIONS)
             
             #print('TEST22222222')
             if landmark[3][0]< landmark[2][0] and landmark[4][0]< landmark[2][0]:
                thumbIsOpen = "true"
                
                
             if landmark[7][1]< landmark[6][1] and landmark[8][1]< landmark[6][1]:
                firstFingerIsOpen = "true" 
                
             if landmark[6][1]< landmark[8][1] and landmark[6][1]< landmark[7][1] and landmark[5][1] < landmark[8][1]:
                finger1 = "true" 
                
             if landmark[10][1]< landmark[12][1] and landmark[10][1]< landmark[11][1] and landmark[9][1] < landmark[12][1]:
                finger2 = "true"
                
             if landmark[14][1]< landmark[16][1] and landmark[14][1]< landmark[15][1] and landmark[13][1] < landmark[16][1]:
                finger3 = "true" 
                
             if landmark[18][1]< landmark[20][1] and landmark[18][1]< landmark[19][1] and landmark[17][1] < landmark[20][1]:
                finger4 = "true"        
                
                
                
             if landmark[11][1]< landmark[10][1] and landmark[12][1]< landmark[10][1]:
                secondFingerIsOpen = "true"  
             if landmark[15][1]< landmark[14][1] and landmark[16][1]< landmark[14][1]:
                thirdFingerIsOpen = "true"
             if landmark[19][1]< landmark[18][1] and landmark[20][1]< landmark[18][1]:
                fourthFingerIsOpen = "true" 
             t= landmark[17][0] +50
             t2= landmark[17][0] -50

             if landmark[9][0]>t2 and landmark[9][0]<t and landmark[13][0]>t2 and landmark[13][0]<t and landmark[5][0]>t2 and landmark[5][0]<t : 
                checkk="true"
                #print("da5alttttt")
             if landmark[5][1]> landmark[8][1] and landmark[5][1]> landmark[7][1] :
                check2 = "true" 
             font = cv2.FONT_HERSHEY_SIMPLEX
             text='null'
             text2='null'
             tto=""
             ca=""
            
             
             if  (thumbIsOpen == "false" and check2 == "true" and secondFingerIsOpen == "false" and thirdFingerIsOpen == "false" and fourthFingerIsOpen == "false" and checkk=="false" ) :
                text="pose 4"
             else:   
            
             
            
                 if  (thumbIsOpen == "false" and firstFingerIsOpen == "true" and secondFingerIsOpen == "true" and thirdFingerIsOpen == "false" and fourthFingerIsOpen == "false" and checkk=="false") :

                    text="pose 2" 
                    imgcanvas=np.zeros((480,640,3),np.uint8)



        #                 tto="Pause"
        #                 player.pause()
                    ori=[]
                
#              if   finger1 == "true" and finger2 == "true" and finger3 == "true" and finger4 == "true" and landmark[8][1] > landmark[3][1] :
#                 text="rock" 
#                 player.pause()
#                 ori=[]
#                 imgcanvas=np.zeros((480,640,3),np.uint8)
             if  checkk == "true" and firstFingerIsOpen == "true" and secondFingerIsOpen == "true" and thirdFingerIsOpen == "true" and fourthFingerIsOpen == "true":
                 text="pose 3"
                 imgcanvas=np.zeros((480,640,3),np.uint8)
                 ori=[]
             else:
                 if  thumbIsOpen == "true" and firstFingerIsOpen == "true" and secondFingerIsOpen == "true" and thirdFingerIsOpen == "true" and fourthFingerIsOpen == "true":
                    text="pose 1"
                    imgcanvas=np.zeros((480,640,3),np.uint8)
#                     player.play()
#                     tto="Play"
    #                 state="stable"
                    order=[]


             if(text=="null"):
                    imgcanvas=np.zeros((480,640,3),np.uint8)
#                     state="stable"

                    order=[]
                    ori=[]
#              dis=DistanceAB(sa11,sa1,landmark[5][0],landmark[5][1])
#              if(sa1<landmark[5][1] and sa2<landmark[9][1] and (dis>3) ):
#                     check3="move"
                    
#              dis=DistanceAB(prevx3,prevy3,landmark[8][0],landmark[8][1])
#              if(text=="scissor" and first=="false" ):
#                     if(prevy3<landmark[8][1] and prevf1y<landmark[12][1] and (dis>6) and check3=="stable" ):
#                         print('TESTTTTTTT')
                    



             dis=DistanceAB(prevx3,prevy3,landmark[8][0],landmark[8][1])
             
             disfi=DistanceAB(prev0,prev01,landmark[0][0],landmark[0][1])
             dis2=abs(landmark[0][0]-prev0)
#              print(landmark[0][2])
             if(text=="pose 2" and first=="false" ):
                if(disfi<5):
                   stableno2=stableno2+1
                if(stableno2>7):
                            
                    tto="Pause"
                    player.pause()
                    stableno2=0
             if(text=="pose 1" and first=="false" ):
                if(disfi<5):
                   stableno3=stableno3+1
                if(stableno3>7):
                            
                    tto="Play"
                    player.play()
                    stableno3=0
             if(text=="pose 4" and first=="false" ):
                    if(dis<5):
                                stableno4=stableno4+1
                    if(stableno4>4):
                            
                        imgcanvas=np.zeros((480,640,3),np.uint8)
#                     state="stable"

                   
                        ori=[]
                        stableno4=0
                    imgcanvas = cv2.line(imgcanvas, (prevx3,prevy3), (landmark[8][0],landmark[8][1]), color, thickness) 
                    ori.append([landmark[8][0],landmark[8][1]])
#                     print(prevx3)
#                     print(landmark[8][0])
#                     print(landmark[8][1])
    
                   
#                     print(angle)
#                     if(angle >= -45 and angle < 45):
# #                         print("right")
#                         di="right"
# #                         if(len(order2)==0):
#                         order2.append(di)

#                     else:
#                        if (angle >= 135 or angle < -135):
# #                          print("left")
#                          di="left"
# #                          if(len(order2)==0):
#                          order2.append(di)
# #                        else:
# #                             di="no"


    #                 if(prevx3<landmark[8][0] and (dis>6)):

    #                     incno2=incno2+1
    #                    # print("inscrese")
    #                     if(incno2>=3):
    #                         state2="inc"
    #                         test_list_set2 = set(order2)
    #                         if state2 in test_list_set2 :
    #                           #print ("Element Exists")
    #                           ti=""
    #                         else:
    #                             if(len(order2)==0):
    #                                 order2.append(state2)
    #                                 incno2=0
    #                                 decno2=0

    #                 else:
    #                       if(prevy3>landmark[8][0] and (dis>6)):

    #                                         decno2=decno2+1
    #                                         #print("decresese")
    #                                         if(decno2>=3):
    #                                            state2="dec"
    #                                            test_list_set2 = set(order2)
    #                                            if state2 in test_list_set2 :
    #                                                ti=""
    #                                               #print ("Element Exists")
    #                                            else:
    #                                               if(len(order2)==0): 
    #                                                 order2.append(state2)
    #                                                 incno2=0
    #                                                 decno2=0
    #                       else:
    #                             state2="stable"







             dis=DistanceAB(prevx2,prevy2,landmark[17][0],landmark[17][1])
             dis=abs(landmark[17][0]-prevx2)
             dis2=abs(landmark[0][0]-prev0)
             
            
                

             if(text=="pose 3" and first=="false"):
                

                if(prevx2<landmark[17][0] and prevx21<landmark[18][0] and prevx22<landmark[19][0] and  prevx23<landmark[20][0] and (dis>10)  ):
#                         print(dis)
#                         print(prevx2)
#                         print(landmark[17][0])
#                             text55="increseee"
#                         incno=incno+1
                            print("inscrese")
#                         print(order)
#                         if(incno>=3):
                            state="inc"
                            test_list_set = set(order)
                            if state in test_list_set :
                              #print ("Element Exists")
                              t=""
                            else:
                                order.append(state)
                                stableno=0
                                print(order)
                else:
                  if(prevx2>landmark[17][0] and prevx21>landmark[18][0] and prevx22>landmark[19][0] and  prevx23>landmark[20][0] and (dis>10)):
#                                     print(dis)
#                                     print(prevx2)
#                                     print(landmark[17][0])
#                                     text55="decrese"
#                                     decno=decno+1
                                       print("decresese")
#                                     print(order)
#                                     if(decno>=3):
                                       state="dec"
                                       test_list_set = set(order)
                                       if state in test_list_set :
                                           t=""
                                          #print ("Element Exists")
                                       else:
                                          order.append(state)
                                          print(order)
                                          stableno=0

                  else:

                        incno=0
                        decno=0
                        state="stable"
                        tto=""
                        stableno=stableno+1
                        
                        if(stableno>5):
                            order=[]
                        if(len(order)>1):
                            print(order)

                            if(order[-1]=="dec"):
                              print("NEXTTTT")
                              player.next()
                              
                              tto="Next"
                            else:
                              if(order[-1]=="inc"):
                                print("PREVVVV")
                                player.previous()
                                tto="Previous"
                            order=[]

             #print(order)
             cv2.putText(image, text, (50, 50), font, 1, (0, 255, 255), 2,cv2.LINE_4)   
#              cv2.putText(image, tto, (50, 150), font, 1, (0, 255, 255), 2,cv2.LINE_4)   
#              cv2.putText(image, tto, (50, 200), font, 1, (0, 255, 255), 2,cv2.LINE_4) 


             color = (0, 255, 0)

             thickness = 3

             
             image= cv2.rectangle(image, (brect[0], brect[1]), (brect[2], brect[3]),(0, 0, 0), 1)
#              if(first=="false"):
#                  image = cv2.line(image, (prevx3,prevy3), (landmark[8][0],landmark[8][1]), color, thickness)








             prev0=landmark[0][0]
             prev01=landmark[0][1]
             prevx2=landmark[17][0]
             prevx21=landmark[18][0] 
             prevx22=landmark[19][0]
             prevx23=landmark[19][0]
             prevy2=landmark[17][1]

             prevx3=landmark[8][0]
             prevy3=landmark[8][1]
             prevf1y=landmark[12][1]
                
             sa1=landmark[5][1]
             sa11=landmark[5][0]
             sa2=landmark[9][1]


             first="false"









        imggray=cv2.cvtColor(imgcanvas,cv2.COLOR_BGR2GRAY)
        _, imginv = cv2.threshold(imggray,50,255,cv2.THRESH_BINARY_INV)
        imginv=cv2.cvtColor(imginv,cv2.COLOR_GRAY2BGR)
#         image=cv2.bitwise_and(image,imginv)
#         image=cv2.cv2.bitwise_or(image,imgcanvas)


        output = image.copy()
        gray = cv2.cvtColor(imgcanvas, cv2.COLOR_BGR2GRAY)
        


    # setting threshold of gray image
        _, threshold = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

        contours, _ = cv2.findContours(threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        cv2.drawContours(imgcanvas, contours, -1, (0, 255, 255), 1)
        

        i = 0

        for contour in contours:

        # here we are ignoring first counter because 
        # findcontour function detects whole image as shape
          if i == 0:
            i = 1
            continue

        # cv2.approxPloyDP() function to approximate the shape
          approx = cv2.approxPolyDP(contour, 0.01 * cv2.arcLength(contour, True), True)
          area = cv2.contourArea(contour)

#           print('areaaa con')
#           print(area)
          
          
          
          
          too="NO"
          ca="noo" 
          
        # putting shape name at center of each shape
          if (len(approx) > 7 ) :
#             print(is_clockwise(ori))
              
#             print(ori)
#             print(approx)
            first2="true"
            
            too="circle"
            
            Final=is_clockwise(ori)
            area = cv2.contourArea(approx)
#             print('areaaa')
#             print(area)
            ori=[]
#             if(area>100):
            if(Final==True):
                ca="Volume up"


                if(v<=100):
                    v=v+10
                    player.setvol(v)
            else:
                if(Final==False):
                     ca="Volume down"
                     v =v - 10
                     player.setvol(v)
            
            print(ca)
#             if(len(order2)>=1):
#             direction(order2)
#                 if(o=="pos"):
#                     ca="clockwise"

#                     v=v+10
#                     if(v<=100):

#                             player.setvol(v)

#                 else:
#                     ca="aniclockwise"
#                     v=v-10
#                     player.setvol(v)
                
            
            
#             if(len(order2)>=1):
#               if(order2[0]=="right"):
#                 ca="clockwise"
                
#                 v=v+10
#                 if(v<=100):
#                     player.setvol(v)
#               else:
#                 ca="aniclockwise"
#                 v=v-10
#                 player.setvol(v)

            imgcanvas=np.zeros((480,640,3),np.uint8)
            order2=[]




#         cv2.putText(image, too, (50, 400), font, 1, (0, 255, 255), 2,cv2.LINE_4) 
#         cv2.putText(image, ca, (50, 200), font, 1, (0, 255, 255), 2,cv2.LINE_4)
#         cv2.putText(image, tto, (50, 150), font, 1, (0, 255, 255), 2,cv2.LINE_4)   
        cv2.imshow('MediaPipe Hands', image)

        cv2.imshow('tt', imgcanvas)












        ret, frame = cap.read()

#         if cv2.waitKey(10) & 0xFF == ord('q'):
#                 break 
        if  event == sg.WIN_CLOSED:
            
            break       
        ret, frame = cap.read()                               # Read image from capture device (camera)
#         imgbytes=cv2.imencode('.png', image)[1].tobytes()     # Convert the image to PNG Bytes
#         window.FindElement('_IMAGE_').Update(data=imgbytes)   # Change the Image Element to show the new image
player.stop()    
cap.release()
cv2.destroyAllWindows()    
```

```python
import PySimpleGUI as sg

#
# An Async Demonstration of a media player
# Uses button images for a super snazzy look
# See how it looks here:
# https://user-images.githubusercontent.com/13696193/43159403-45c9726e-8f50-11e8-9da0-0d272e20c579.jpg
#
def MediaPlayerGUI():
    # Set the backgrounds the same as the background on the buttons
    # Images are located in a subfolder in the Demo Media Player.py folder
    image_pause = 'C:\\Users\\farah\\Desktop\\music\\play.png'
    image_prev = 'C:\\Users\\farah\\Desktop\\music\\prevoff.png'
    image_next = 'C:\\Users\\farah\\Desktop\\music\\nextoff.png'
    

    # Use the theme APIs to set the buttons to blend with background
    sg.theme_button_color((sg.theme_background_color(), sg.theme_background_color()))
    sg.theme_border_width(0)        # make all element flat

    # define layout of the rows
    layout= [
             [sg.Button(image_filename=image_prev, image_size=(100, 100), image_subsample=2,  key='PREV'),
              
              sg.Button(image_filename=image_pause, image_size=(100, 100), image_subsample=2,  key='PAUSE'),
              
              sg.Button(image_filename=image_next, image_size=(100, 100), image_subsample=2,  key='NEXT'),
             
            ]
             ]

    # Open a form, note that context manager can't be used generally speaking for async forms
    window = sg.Window('Media File Player', layout, default_element_size=(20, 1), font=("Helvetica", 25))
    # Our event loop
    while True:
        event, values = window.read(timeout=100)        # Poll every 100 ms
        if event == sg.WIN_CLOSED:
            break
        # If a button was pressed, display it on the GUI by updating the text element
        if event != sg.TIMEOUT_KEY:
            window['-OUTPUT-'].update(event)

MediaPlayerGUI()
```

```python

import cv2
import mediapipe as mp
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands

# For static images:
IMAGE_FILES = []
with mp_hands.Hands(
   
    max_num_hands=2,
    min_detection_confidence=0.5) as hands:
  for idx, file in enumerate(IMAGE_FILES):
    # Read an image, flip it around y-axis for correct handedness output (see
    # above).
    image = cv2.flip(cv2.imread(file), 1)
    # Convert the BGR image to RGB before processing.
    results = hands.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

    # Print handedness and draw hand landmarks on the image.
    print('Handedness:', results.multi_handedness)
    if not results.multi_hand_landmarks:
      continue
    image_height, image_width= image.shape
    annotated_image = image.copy()
    for hand_landmarks in results.multi_hand_landmarks:
      print('hand_landmarks:', hand_landmarks)
      print(
          f'Index finger tip coordinates: (',
          f'{hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].x * image_width}, '
          f'{hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].y * image_height})'
      )
      mp_drawing.draw_landmarks(
          annotated_image, hand_landmarks, mp_hands.HAND_CONNECTIONS)
    cv2.imwrite(
        '/tmp/annotated_image' + str(idx) + '.png', cv2.flip(annotated_image, 1))

# For webcam input:
cap = cv2.VideoCapture(0)
with mp_hands.Hands(
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5) as hands:
  while cap.isOpened():
    success, image = cap.read()
    if not success:
      print("Ignoring empty camera frame.")
      # If loading a video, use 'break' instead of 'continue'.
      continue

    # Flip the image horizontally for a later selfie-view display, and convert
    # the BGR image to RGB.
    image = cv2.cvtColor(cv2.flip(image, 1), cv2.COLOR_BGR2RGB)
    # To improve performance, optionally mark the image as not writeable to
    # pass by reference.
    image.flags.writeable = False
    results = hands.process(image)

    # Draw the hand annotations on the image.
    image.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    if results.multi_hand_landmarks:
      for hand_landmarks in results.multi_hand_landmarks:
        mp_drawing.draw_landmarks(
            image, hand_landmarks, mp_hands.HAND_CONNECTIONS)
        x = [landmark.x for landmark in hand_landmark.landmark]
        y = [landmark.y for landmark in hand_landmark.landmark]
        center = np.array([np.mean(x)*image_width, np.mean(y)*image_hight]).astype('int32')
        cv2.circle(image, tuple(center), 10, (255,0,0), 1)  #for checking the center 
        cv2.rectangle(image, (center[0]-200,center[1]-200), (center[0]+200,center[1]+200), (255,0,0), 1)
    cv2.imshow('MediaPipe Hands', image)
    if cv2.waitKey(5) & 0xFF == 27:
      break
cap.release()
```
